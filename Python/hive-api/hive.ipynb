```python
from pyspark.sql import SparkSession

def nested_f(x):
    return x + 1

print(1)
spark = SparkSession.builder.appName("test_udf").enableHiveSupport().getOrCreate()
print(1)
sc = spark.sparkContext
print(1)
sc.setLogLevel("WARN")
print(1)
log4jLogger = sc._jvm.org.apache.log4j
print(1)
LOGGER = log4jLogger.LogManager.getLogger(__name__)
print(1)
LOGGER.warn("logger before run...")
print(1)
# register udf
spark.udf.register("test_udf", nested_f, 'int')
print(1)
# use udf in sql
spark.sql("select 3").show()
print(1)
LOGGER.warn("logger after run...")
print(1)
spark.stop()
```
