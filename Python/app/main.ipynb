```python

# import necessary libraries
import os
import glob
  
  
# use glob to get all the csv files
# in the folder
# path = r'C:\Users\Administrator\Downloads\testing_file'
path = os.path.join(os.getcwd(), ".data")

csv_files = glob.glob(os.path.join(path, "*.xlsx"))

df = pd.DataFrame()  
# loop over the list of csv files
for f in csv_files:
    
    # read the csv file
    read = pd.read_excel(f, header= 1)
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    df = pd.concat([df,read])
```

```python
def read_sub_file(file_path):
    df = pd.read_excel(file_path)    
    unpivot = pd.melt(df, id_vars=['Referrence / key ', 'Context ', 'Char. Limit ', 'EN Source '], value_vars=['af_za',
        'am', 'ar', 'az_latn', 'bg', 'bn', 'ca', 'cs_cz', 'da_dk', 'de', 'el',
        'es', ' es_419', 'es_ar', 'es_mx', 'et', 'fa', 'fi_fi', ' tl_ph', 'fr',
        'ga', 'hr', 'he', 'hi', 'hu', 'hy', 'id', 'it', 'ja', 'ka', 'kk', 'km',
        'ko', 'lt', 'lv', 'mk', 'ml', 'mn_mn', 'ro_md', 'ms', 'my', 'nb_no',
        'nl_nl', 'pl', 'pt', 'pt_br', 'ro', 'ru', 'sk', 'sl', 'sq', 'sr_cyrl',
        'sr_latn', 'sv', 'sw', 'th', 'tr', 'uk', 'ur', 'uz', 'vi', 'zh_cn',
        'zh_tw'],
            var_name='translate_to', value_name='content')

    return unpivot.loc[~unpivot.content.isna() , :]


csv_files = glob.glob(os.path.join(path, "*.xlsx"))
full_df = pd.DataFrame()

for file_path in csv_files:
    read = read_sub_file(file_path)
    full_df = pd.concat([full_df,read])

full_df.drop_duplicates(inplace=True)
full_df
```

```python
full_df['content'] = full_df['content'].fillna('')
full_df['Context '] = full_df['Context '].fillna('')
full_df['Referrence / key '] = full_df['Referrence / key '].fillna('')
full_df['Char. Limit '] = full_df['Char. Limit '].fillna('')

pivot = pd.pivot_table(full_df, index=['Char. Limit ','EN Source '],
                       columns=['translate_to'], 
                       values='content',
                       aggfunc=lambda x: ' '.join(x)
                       )

df_export = pivot.reset_index().fillna('')
df_export
```
